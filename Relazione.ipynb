## INTRODUCTION

A Plummer Sphere of total mass $M$ and scale radius $b$ is a gravitationally bound system, with some peculiar properties. Given a set of spherical coordinates centered on the sphere, it is described by the following potential-density pair:

$$ \rho(r) = \frac{ 3M }{ 4\pi b^{3} } \Bigl({ \frac{1}{ 1 + \frac{{r}^{2}}{{b}^{2}} } }\Bigr)^\frac{5}{2} \;\;\;\;\;\; \rightarrow \;\;\;\;\;\; \nabla^{2}(\Phi) = 4\pi G \rho \;\;\;\;\;\; \rightarrow  \;\;\;\;\;\; \Phi(r) = \frac{ -GM }{ \sqrt{r^{2}+b^{2}} } $$

Here are some main observations one can make, by looking ad $\rho$:
> The system is isotropic in $\theta$ and $\phi$ 
> The density is constant in the centre, and drops quickly for r >> b <br>
> The total mass M of the system is finite: integrating $ 4\pi r^{2}dr $ from R = 0 to R = $\infty$ does not cause M(R) to diverge: instead, it reaches its maximum value M.

The equation of $\rho$ can be plugged into Poisson's equation to provide the expression of $\Phi$.
> There are no discontinuities in the centre of the sphere: $\Phi(0) = - \frac {GM}{B} $ <br>
> For larger values of $r$, $\Phi$ resembles a keplerian potential.

The objective of this work is to define a self-gravitating system at the equilibrium: if one generates a plummer sphere with every particle velocity set to zero, it will inevitably collapse towards its central point. The purpose of this section is to look for a velocity distribution, combined with a spatial distribution of particles, which is going to prevent the gravitational collapse of the system.

We define the _Distribution Function f_, as the probability density $f(x,x+dx;v,v+dv)$. It is a 6-dimensional function, whose integral provides the probability of finding a particle between x and x+dx, v and v+dv, in its 6-dimentional phase space. Given a potential $\Phi$, one can use Newton's second law and solve the equation of motion for every particle. Is there a relationship between the gravitational potential and the distribution function?

Let's start with an important assumption: _THE NUMBER OF PARTICLES IS CONSERVED_ <br><br>
There is a straightforward consequence from this: the derivative of f in time, $\frac{df}{dt}$, is zero. Let's define $w$ = ( x, v ) and its derivative in time $\dot{w}$ = ( v, a ) : the derivative above ( also known as _Material Derivative_ ) leads to the following result:

$$ \frac{\partial f}{\partial t} = - \frac{\partial ( f \dot{w}_{i})}{\partial w_{i}} $$

where _repeated indeces are summed_, and i = 1, 2,... 6 = $x, y, z, v_{x}, v_{y}, v_{z}$. 

> Let's assume that _the interaction is purely gravitational_: <br>

The x and v variables are independent. With the above assumption, the right hand side becomes $ - \dot{w}_{i} \frac{\partial f}{\partial w_{i}} $ 

This leads to the _Collisionless Boltzmann Equation, CBE_, in one of its forms: <br>
> $$ \frac{\partial f}{\partial t} + \dot{w}_{i} \frac{\partial f}{\partial w_{i}} = 0 $$

Let's rewrite the solution in a more "physical" way, by explicitly showing the dependence from $\Phi$:
> $$ \frac{\partial f}{\partial t} + v \cdot \nabla f - \nabla \Phi \cdot \frac{\partial f}{\partial v} = 0 $$

###### If one can solve the equation for f given $\Phi$, a distribution of positions and velocities can be found and it is a stationary solution.

Solving this equation for a system of particles is _HARD_, therefore one can work to find an "averaged version" of the _CBE_. The first step is to define the a new density $\nu$ = $\int{f d^{3}v }$ and to integrate the CBE in dv.

An intermediate result comes from the definition of $\overline{v_{i}} = \frac{1}{\nu} \int{f v_{i} d^{3}v }$ and gives the _Averaged Continuity Equation_, which is similar to the continuity equation above: 

> $$ \frac{\partial \nu}{\partial t} + \sum_{i}{( \nu \; \overline{v_{i}} )} = 0 $$

Multyplying the CBE equation by $v_{j}$ and then integrating in $d^{3}v$ leads to the _Averaged Equation of Motion_:

> $$ \nu \frac{\partial{ \overline{v_{j} }}} {\partial {t}} + \nu \;\overline{v_{i}} \frac{\partial{\overline{v_{j} } } }{\partial{x_{i}}} = - \nu \frac{\partial{ \Phi } }{\partial{x_{i}}} - \frac{\partial{( \nu \sigma^{2}_{ij} )} }{\partial{x_{i}}} $$

where i,j = 1,2,3 and i is summed everywhere and $ \sigma^{2}_{ij} = \overline{v_{i}v_{j}} - \overline{v_{i}}\;\overline{v_{j}} $ is the velocity dispersion.

Further integration in $d^{3}x$ brings to an _Averaged Conservation of Energy Equation_:

$$ \frac{1}{2} \frac{\partial ^{2}}{\partial t^{2}} I_{jk} = 2 T_{j}{k} + \Pi{j}{k} + W_{jk} $$

Where I, T and W are respectively the Inertia, Kinetic Energy and Potential Energy tensors, and $\Pi_{jk} = \int{\rho \sigma^{2}_{jk}d^{3}x } $

##### By averaging an exact solution of the system, one can still follow the process above and recover some average informations on the evolution of the system.

Let's look for a specific solution of a physical system, and start by stating _Jeans's Theorem_:

> If I = I(x,v) is an integral of motion of the system $\rightarrow$ I is a steady-state solution of the CBE <br>

> If f is a steady-state solution of the CBE $\rightarrow$ f is an integral of motion of the system.

Another powerful theorem is the following ( by Lynden-Bell ):

> Given a spherically symmetrical system with an isotropic velocity distribution, one can always find a CBE solution f as a combination of _two_ integrals of motion, which are the total energy E and the angular momentum |L|

One can now focus on looking for a specific distribution f, which is taken to be only dependent on E.

The following fundamental quantities are defined:
$$ Relative\;Potential \rightarrow \psi = -\Phi + \Phi_{0} \;\;\; Relative\;Energy \rightarrow \epsilon = -E + \Phi_{0} = \psi - \frac{v^{2}}{2} $$

$\Phi_{0}$ is defined to satisfy the following conditions: f > 0 if $\epsilon$ > 0 , and f = 0 if $\epsilon$ $\leq$ 0 $\rightarrow$ one can narrow down their work to bound particles.

Since the distribution is spherically symmetrical, one can easily verify with their definition that, being $\overline{v^{2}_{r}} = \overline{v^{2}_{\theta}} = \overline{v^{2}_{\phi}} $, $\sigma^{2}_{ij} = 0 $ 

The next step is to plug $\Psi$ into Poisson's Equation ( in spherical coordinates ) and to redefine $\rho$ as $\int{f(\epsilon)d^{3}v}$. There is no dependence on $\theta$ or $\phi$, therefore one comes to the following, fundamental result:

$$ \frac{1}{r^{2}}\frac{\partial}{\partial{r}}r^{2}\frac{\partial}{\partial{r}}\psi = -4\pi G \int^{\sqrt{2\psi}}_{0}{4\pi v^{2}f\Bigl(\psi - \frac{v^{2}}{2}\Bigr)dv}$$

Notice the upper limit of integration: if $v \geq \sqrt{2\psi}$, the particle is free.

A subclass of solutions to this equation comes from the so-called _polytropes_: this names comes from the branch of physics of fluids. In these systems, the pressure $p$ and its density $\rho$ are correlated through a relation of the form $p = K \rho^{\alpha}$, where K is a proportionality constant and $\alpha$ is the _polytropic index_.

Let's assume the following form for f($\epsilon$):
> $f(\epsilon) = F \epsilon^{n-\frac{3}{2}} $ if $\epsilon \geq 0$ <br><br>
> $f(\epsilon) = 0 $ if $\epsilon \leq 0$

A change of variables and the request that $\int{f} < \infty$ leads to the final result $\rho = c_{n}\psi^{n}, n \geq \frac{1}{2}$. Let's $\rho$ into Poisson's Equation: one can quickly come to the following expression:

$$ \frac{1}{r^{2}}\frac{\partial}{\partial{r}}r^{2}\frac{\partial}{\partial{r}}\psi + 4\pi G c_{n}\psi^{n} = 0 $$

Another change of variables with $ s = \frac{r}{r_{0}}, \Psi = \frac{\psi}{\psi_{0}}$ leads to the _Lane-Emden Equation_:
$$ \frac{1}{s^{2}}\frac{\partial}{\partial{s}}s^{2}\frac{\partial}{\partial{s}}\Psi = \Psi^{n} $$ 
if $ \Psi > 0 $
<br>
$$ \frac{1}{s^{2}}\frac{\partial}{\partial{s}}s^{2}\frac{\partial}{\partial{s}}\Psi = 0 $$ 
if $ \Psi \leq 0 $
<br>
One can set the boundary conditions for the equation: $\Psi(0) = 1, \Psi(\infty) = 0, \dot{\Psi}(0) = 0 $ and look for a solution of the system.

Said solution has been found for polytropic gases, but can be used to describe a stellar system as well:

$$ \Psi = \frac{1}{\Bigl(1+\gamma s^{\alpha}\Bigr)^{\beta}} \;\;;\;\; \frac{d\Psi}{ds} = \frac{- \beta \gamma \alpha s^{\alpha -1 }}{\Bigl( 1 + \gamma s^{\alpha} \Bigr)^{\beta + 1}}$$

One can plug $\Psi$ into the Lane-Emden equation and look for the values of $\alpha, \beta$ and $\gamma$ ( and n ) that satisfy the equality, i.e. the right hand-side is equal to the left hand-side:

$$ \alpha = 2 \;\;\; \beta = \frac{1}{2} \;\;\; \gamma = \frac{1}{3} \;\;\; n = 5 $$

_By plugging these values into the expression of $\it{\Psi}$, one can find an expression similar to the potential-density pair of a Plummer Sphere!_ 
##### One can therefore conclude that the system under study in this work provides a polytropic solution to the Lane-Emden equation: therefore, one can find the distribution function, which will provide the velocity distribution that needed to prevent the gravitational collapse of the system.

## METHODS

To simulate the evolution in time of a Plummer Sphere of radius $b$ and total mass $M$ in a numerical simulation, the first step is to generate a set of $N$ identical particles in 3D space to mirrors the given density profile: to do that, one can start with $\rho$: where is it most likely that a particle will be found/placed? It is easy to integrate $\rho(r)$ and to come to an expression of M(R): 

$$ M(R) = \int^{R}_{0}{4 \pi r^{2}\rho(r)dr } = \frac{M R^{3}}{b^{3}}\Bigl(1+\frac{R^{2}}{B^{2}}\Bigr)^\frac{-3}{2} $$

$ \frac{M(R)}{M}$ can be seen as a ( cumulative ) probability distribution p(R): given a random radius R, one knows the fraction of mass enclosed in that radius. One now needs to invert this equality to find an expression from R(p):

$$ R(p) = b \frac{p^\frac{1}{3}} {\sqrt{1-p^\frac{2}{3}}} $$

The process used to to compute the distance from the centre of the system is the following: one extracts a random number from a uniform probability ditribution between 0. and 1., and computes R.

A similar process is performed to compute the angles $\theta$ and $\phi$: since the system is spherically symmetrical, one can come to the following results:

$$ \theta(p) = arccos( 1 -2p ) \;\;\;\;\; \phi(p) = 2\pi p $$

The same algorithm is now used to extract the values of $\theta$ and $\phi$: a random number is extracted between 0. and 1. from a uniform probability distribution _for each independent coordinate_. Given a vector $( r, \theta, \phi )$ one can easily compute its ( x, y, z ) counterpart and place the particle in space.



It would be nice to perform something similar to find a velocity vector for a particle: this is not possible, as one cannot "flip" the integral relationship between p($\epsilon$) and find $\epsilon(p)$. There is no analytical solution, but one can sample the f($\epsilon$) distribution with a MonteCarlo Rejection Method. Here are the steps to follow to perform the sampling:

> Once a particle is set in space, its potential $\phi$ and its relative potential $\psi$ are set.<br><br>
> The substitution q = $\frac{v}{\sqrt{2\psi}}$ is performed.
> $f(\epsilon) = F \epsilon^{n-\frac{3}{2}} $ becomes $f(q) = K q^{2}\Bigl(1-q^{2}\Bigr)^\frac{7}{2}$, where K is a proportionality constant: let's put it equal to 1, as it will soon be clear that it is not relevant

One now has to extract two random numbers:
> q, which is extracted from a uniform probability distribution between 0. and 1.
> y, which is extracted from a uniform probability distribution between 0. and the max( f(q) ).
> if y < f(q), q is sampled: v, the magnitude of the velocity, is set: $v = \sqrt{2\psi}q$ 
> if y > f(q), q is rejected: _a new pair of variables is extracted_, and the sampling continues.
> $\theta$ and $\phi$ are independently extracted from their respective equations above, since the velocity distribution is isotropic as well.
> ( vx, vy, vz ) is computed and assigned to the particle.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from matplotlib import rcParams
rcParams.update({'figure.autolayout': True})

from scipy.optimize import curve_fit
def f(x):
    return x*x*( 1-x*x )**7/2
x = np.linspace(0.,1., 500)
y = f(x)
plt.plot(x,y, label='f(q)')
plt.legend()
plt.show()
```

_As this image shows, what matters when sampling with the MC Rejection method is the shape of the distribution: if one extracts a random number between 0 and the maximum of a function, the function has a narrow peak and q is extracted far from the peak, one will likely have to discard the pair ( q, y ). Another ( q, y ) pair will be extracted, and it will be more likely that it will be accepted if q is near the peak of the curve. 
The more particles are extracted, the more the distribution of generated q values will mirror the one they have been sampled from._

Lastly, one has to perform the numerical simulation: a treecode program by J. Barnes is used to evolve the system: a C++ program performs the previously described algorithm to generate the initial conditions and to read the output file. The initial conditions are written in the treecode_input.txt file, in the required format. The treecode output is written in treecode_output.txt and treecode_log.txt. a jupyter notebook grabs the treecode output, elaborated by the C++ program and archived in /my_output/data/, and plots the graphs.

> In this setup, the total mass of the system has been set to M = 100 Solar Masses = 100 i.u., and the Scale Radius b of the system has been set to 9 Parsec = 9 i.u.
> N has been set to 100'000 particles.

Now, the parameters:
> "exec ./treecode/treecode treecode_input.txt treecode_output.txt options=out-phi dtime=1/512 eps=1/256 theta=1/8 tstop=4 dtout=1/16 > treecode_log.txt";

This is the command line found in system.cpp which starts the treecode.

> dtime is the fixed time interval at which the system evolves with its leapfrog method: since the dynamical time is 3 and it has been recommended to set dtime to be at least a thousand times smaller, 1/512 can be safely used. In the results portrayed in this work, dtime has been set to 1/1024 instead.

> eps is the gravitational softening parameter, and it is set to 1/256. 1/512 would probably be better, as it was recommended to use a fraction of $\sqrt{\frac{Volume}{N}}$, but one can personally be satisfied with these results.

> theta is the accuracy parameter: the smaller theta, the longer the calculations. Since the particle resolution is really small, a "big" value is chosen for theta, set to 1/8

> the dynamical time of the system is roughly calculated as $\frac{b\sqrt{b}}{\sqrt{GM}} \sim 3$ in internal units, therefore the system is reasonably stopped at tstop = 4.

> lastly, dtout is the frequency of output in treecode_output.txt: one must avoid big values of dtout with lots of particles, to avoid the creation of gargantuan-sized output files. dtout = 1/32 has been used in this work, but 1/16 should still fill your folder with lots of MB's of data.

### Density Profile

Lastly, the hardest result to obtain: the density profile. To determine the value of $\rho(r)$, each particle has been sorted, from the closest to the centre to the farthest to the centre. An algorithm started counting the number of particles between a radial value r, and r+dr: the total mass enclosed in that interval, equal to $count*m_{i}$ in the code, is then divided by the volume of its spherical shell, $4\pi r^{2} dr$. 

> The computation of dr has been performed both with a logarithmic expansion of dr and a fixed dr, and performs better in the former case. The fixed dr method is now disabled, but it can still be enabled in the code. Progressively enlarging dr helps the sampling of the distribution, while a fixed value of dr needs an accurate choice of its value, which changes drastically the shape of the plot at smaller radii. Since it has been said that the average distance between two particles can be approximately equal to $d = \sqrt[3]{\frac{V}{N}}$, I personally chose to set $V \sim B^{3}$ and to take ten times the value of d: the idea has been "to enclose "a bunch of particles" in a single shell", and it works nicely. However, the "best" value of d changes drastically with the number of particles, and so far no quantitative explanation for this occurrence can be provided.

> It is hard to compute the values of $\rho$ at smaller radii: the system both lacks particles and has small-sized volumes, and provides terrible results at smaller values of r. The lack of a particle may cause the mass over volume ratio to skyrocket, while a total lack of particles in a shell sets $\rho$ to zero most of the time. I personally chose to cut the zero values of $\rho$, to provide a more relevant plot. 

Additional plots can be found in the folder my_output/density/. It can be said that the density profile is nicely mirrored for larger values of r, while its profile falls apart when one gets closer to the centre.

## CONCLUSIONS

A Plummer Sphere has been created and it has not collapsed with the given density profile, as expected.

The system is not an ideal one, being composed of a finite number of particles: sampling a distribution with a MonteCarlo method is feasible, but one has to pay a computational cost price to sit down and perform a lengthy numerical simulation. In this work, relevant issues have been found in generating particles close to the centre of the sphere and at higher speeds: the system is perfectly described as a whole and far away from these "critical" values, however one can find the deviations from the model in its practical, "finite" nature.

To sum up the results:
> The energy of the system is conserved, and the center of mass is stationary: the simulation has been performed successfully.

> The distribution of mass is unchanged in the evolution of the system: while particles can move inwards and outwards, the average mass within a fixed radius remains unchanged.

> There are significant fluctuations in the number of particles sitting at smaller radii: this is due to MonteCarlo methods, random number generation in C++ and the shape of R(p).

> The velocity distribution tail lacks some particles in its tail: this is attributed to the lack of particles near the centre of the system, and the randomity in the value of q.

> It is difficult to probe accurate and precise values of $\rho$ at smaller radii, as the lack of particles causes issues in the calculation of the local density. A variable value of dr ( in logscale ) turns useful to mitigate this issue, making the result more accurate, but not as precise as one may want it to be.

In conclusion, the experimental results accurately confirm the initial hypothesis, with some small, relevant caveats: 

> MonteCarlo methods "backfire" when one wants to accurately sample "singular" points in "pathological" distributions: the generated system is not perfect, both at smaller and bigger radii, but NO complaints can be made inbetween these values.

> The higher the number of particles generated, the better the MC results will be, with *way* longer computational costs.
